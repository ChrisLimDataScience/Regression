## Problem 1

We will be using a subset of the mtcars.csv dataset to perform data abstraction, slicing, dicing and basic analysis using Pandas DataFrame. The dataset is given below.

```
data = {'car name':["Mazda RX4","Mazda RX4 Wag",
                    "Datsun 710", "Hornet 4 Drive",
                    "Hornet Sportabout", "Valiant",
                    "Duster 360", "Merc 240D",
                    "Merc 230", "Merc 280"],
        "Cylinders": [6,6,4,6,8,6,8,4,4,6],
        "Gear": [4,4,4,3,3,3,3,4,4,4],
        "Horsepower": [110,110,93,110,175,105,245,62,95,123],
        "Miles per Gallon":[21.0,21.0,22.8, 21.4,18.7,18.1,
                            14.3,24.4,22.8, 19.2]}
```

1.1 Create a DataFrame using the columns: Cylinders, Gear, Horsepower, Miles Per Gallon. The row index labels are shown below. Print the dataframe.

image.png

```
# Type solution here
import pandas as pd;
df = pd.DataFrame(data=data, columns=["Cylinders","Gear","Horsepower", "Miles per Gallon", "car name"])
df = df.set_index('car name')
print(df)
```

1.2 Print the dataframe in descending order of horsepower followed by ascending order of Miles Per Gallon.

```
# Type solution here
print(df.sort_values(by=['Horsepower','Miles per Gallon'], ascending=[False,True]))
```

1.3 Print the dataframe showing only 4 geared cars.

```
# Type solution here
print(df[df["Gear"]==4])
```

1.4 Print the number of cars that have more than 20 Miles Per Gallon

```
# Type solution here
print(df[df["Miles per Gallon"]>=20])
```

1.5 Now create a figure with 2x2 subplots.

  -  Subplot 1: Create a scatter plot of Horsepower vs Miles per gallon

  -  Subplot 2: Create a histogram of Cylinders

  - Subplot 3: Create a column chart of Gear vs Car Name

  - Subplot 4: Create a box plot of Horsepower

  image.png

```
# Type solution here
import matplotlib.pyplot as plt
import seaborn as sns
fig,ax = plt.subplots(2,2)

ax[0, 0].scatter(df["Horsepower"], df["Miles per Gallon"])
ax[0, 0].set_xlabel("Horsepower")
ax[0, 0].set_ylabel("Miles per Gallon")
ax[0, 1].hist(df["Cylinders"])
ax[0, 1].set_xlabel("Cylinders")
ax[0, 1].set_ylabel("Frequency")
ax[1, 0].bar(df.index, df["Gear"])
ax[1, 0].set_xlabel("Car")
ax[1, 0].set_xticklabels(df.index, rotation = 90)
ax[1, 0].set_ylabel("Gears")

ax[1,1] = sns.boxplot(y=df["Horsepower"])

plt.show()
```

```
# Problem 2

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#2.1
data = pd.read_csv('student_scores.csv')

#2.2
print(data.shape)

#2.3
print(data.head())

#2.4
print(data.describe())

#2.5
data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
# we can clearly see that there is a positive linear relation between the number of hours studied and percentage of score.

#2.6
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

#2.7
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#2.8
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

#2.9
print(regressor.intercept_)

#2.10
print(regressor.coef_)

#2.11
y_pred = regressor.predict(X_test)

#2.12
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df)

#2.13
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))



```
